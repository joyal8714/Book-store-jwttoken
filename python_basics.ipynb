#FUNCTIONS
A function is a block of reusable code that performs a specific task.
Reusability – write once, use many times

Modularity – breaks code into smaller parts

Easy to debug – easier to find errors

Makes code readable

```python
def greet(name="Friend"):
    print("Hello,", name)

greet("Joyal")
greet()   # uses default value
# Think of a function as a machine:

# You give it some input (ingredients) → it does something → it gives output.

# In our example, the function is greet.

# Input: name

# Job: print "Hello, name"

# Output: message on the screen
```
